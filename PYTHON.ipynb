## Creating Frozen class attributes

```python
try:
    from dataclasses import dataclass
    @dataclass(frozen=True)
    class Animal():
        name : str

    dog = Animal('Peter')
    print(dog.name)
    #dog.name='Matt'
    print(dog.__dataclass_fields__,'\n',dog.__dataclass_params__)

except Exception as error:
    print('We have encountered the following error :- ',str(error))    
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate two lists of size 20
x = np.linspace(0, 10, 20)
y1 = np.sin(x)
y2 = np.cos(x)

# Line plot
plt.figure()
plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot')
plt.legend()

# Scatter plot
plt.figure()
plt.scatter(x, y1, label='y1')
plt.scatter(x, y2, label='y2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.legend()

# Bar plot
plt.figure()
plt.bar(x, y1, label='y1')
plt.bar(x, y2, label='y2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar Plot')
plt.legend()

# Histogram
plt.figure()
plt.hist(y1, label='y2')
plt.hist(y2, label='y2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()

# Box plot
plt.figure()
plt.boxplot([y1, y2], labels=['y1', 'y2'])
plt.ylabel('Value')
plt.title('Box Plot')

#scatter plot
plt.figure()
sizes = [max(0, sum(y1)), max(0, sum(y2))]  # Ensure non-negative sizes
labels = ['y1', 'y2']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')

plt.show()
```
