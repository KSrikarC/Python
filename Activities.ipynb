```python
class DataVisual():
    def lineplot(self,x,y, title= 'Output',xl= 'x-axis', yl= 'y-axis'):
        import matplotlib.pyplot as plt
        plt.figure
        plt.plot(x,y)
        plt.xlabel(xl)
        plt.ylabel(yl)
        plt.title(title)
        plt.show()
```

```python
obj = DataVisual()
obj.lineplot([1,2,3],[4,5,6])

```
