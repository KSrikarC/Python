```python
s='abcd'
substrings={}
for i in range(len(s)):
    for j in range(i+1, len(s)+1):
        substr = ''.join(sorted(s[i:j]))
        if substr in substrings:
            substrings[substr] += 1
        else:
            substrings[substr] = 1
count = 0            
for key in substrings:
    n = substrings[key]
    count += (n * (n-1)) // 2
count    
```

```python
#SLIDING WINDOW
def Sliding_Window_Subarrays(arr,length):
    new_arr=[]
    #length = 4
    n = len(arr)
    for i in range(n - length + 1):
        yield(arr[i:i+length])  

def birthday(s,d,m):
    sub_arrays = list(Sliding_Window_Subarrays(s,m))
    count=0
    for _ in sub_arrays:
        if sum(_) == d:
            count+=1
    return count
birthday(s,4,2)    
```

```python
def pageCount(n,p):
    st_curr_page = 1
    end_curr_page = n
    fw_turns = 0
    bk_turns = 0
    while True:
        fw_turns += 1
        st_curr_page += 1
        if st_curr_page == p or st_curr_page+1 == p:
            break

    while True:    
        if end_curr_page%2==0 and end_curr_page == p:
            break
        if end_curr_page%2!=0 and end_curr_page == p or end_curr_page -1 == p:
            break
        bk_turns+=1
        end_curr_page -= 1
    return min(fw_turns, bk_turns)    

print(pageCount(6,2))

def pageCount(n,p):
    return(min(p//2, n//2-p//2))
```
