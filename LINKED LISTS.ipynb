```python
class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        
        
    def print(self):
        if self.head is None:
            print("Linked list is empty")
            return
        
        itr = self.head
        llstr = ''
        
        while itr:
            llstr += str(itr.data)+' --> '
            itr = itr.next
        print(llstr)
        
    def len(self):
        c=0
        itr=self.head
        while itr:
            c+=1
            itr = itr.next
        return c
    
    def insert_at_beginning(self,data):
        node=Node(data,self.head)
        self.head = node
    
    def insert_at_end(self,data):
        if self.head is None:
            node=Node(data,None)
            self.head = node
            return
        
        itr = self.head
        while itr.next:
            itr = itr.next
            
        itr.next = Node(data)             
            
            
    def insert_at(self,data,index):
        if index < 0 or index >= self.len():
              raise Exception("Invalid Index Value")
        if index==0:
            self.insert_at_beginning(data)
            return
        
        itr=self.head
        c=0
        while True:
            if c==index-1:
                itr.next=Node(data, itr.next)
                break
            
            itr = itr.next
            c += 1
        
    def remove_at(self,index):
        if index < 0 or index >= self.len():
            raise Exception("Invalid Index Value")

        if index==0:
            self.head = self.head.next
            return
        
        itr,c=self.head,0
        while True:
            if c==index-1:
                itr.next=itr.next.next
                break
            itr = itr.next
            c+=1
            
    def insert_values(self,data_list):
        self.head = None
        for data in data_list:
            self.insert_at_end(data)
            
    
    
    def remove_by_value(self,data):
        if self.head is None:
            return
        
        flag=False
        itr=self.head
        while itr:
            if itr.data == data:
                flag=True
                break
            else:
                flag=False
            itr=itr.next
        
        if flag:
            if self.head.data==data:
                self.head = self.head.next
                return

            itr = self.head
            c=0
            while 1:
                if itr.next.data==data:
                    itr.next = itr.next.next
                    break

                if itr.next is None:
                    break

                itr = itr.next
                c+=1
        
        else:
            print(data ," not found")
                      
            
            
            
obj = LinkedList()
obj.insert_values([0,1,2,3,4,5])
obj.insert_at_beginning(3)
obj.insert_at_end(5)
obj.insert_at_beginning(2)
obj.insert_at_beginning(1)
obj.insert_at('6',5)
obj.remove_at(2)
obj.print()
obj.remove_by_value(10)
obj.print()
        
```
